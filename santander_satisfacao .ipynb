```python
import pandas as pd
dados_treino = pd.read_csv('train.csv')
dados_teste = pd.read_csv('test.csv')
```

```python
df_treino = dados_treino[['ID', 'var3', 'var15', 'var38', 'TARGET']]
```

```python
df_treino.head(5)
```

```python
df_teste = dados_teste[['ID', 'var3', 'var15', 'var38']]
```

```python
df_teste.head(5)
```

```python
df_treino.corr()
```

```python
atributos = ['ID', 'var3', 'var15', 'var38']
previsto = ['TARGET']
```

```python
X = df_treino[atributos].values
Y = df_treino[previsto].values
```

```python
X
```

```python
Y
```

```python
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

num_folds = 10
seed = 7

kfold = KFold(num_folds, True, random_state = seed)

modelo = LogisticRegression()

resultado = cross_val_score(modelo, X, Y, cv = kfold, scoring = 'accuracy')

print("Acurácia: %.3f" % (resultado.mean() * 100))
```

```python
modelo.fit(X, Y)
modelo.score(X, Y)

print('Coefficient: \n', modelo.coef_)
print('Intercept: \n', modelo.intercept_)


valores_previstos = modelo.predict(df_teste)
```
